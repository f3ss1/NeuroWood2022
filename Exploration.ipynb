```python
import pandas as pd
from matplotlib import pyplot as plt
import os
```

```python
view_dict = {
    0 : os.listdir('Data/Train/drova')[0],
    1 : os.listdir('Data/Train/1')[4],
    3 : os.listdir('Data/Train/3')[0]
    
}
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(plt.imread(f'Data/Train/1/{view_dict[1]}'))
plt.title('1')

plt.subplot(1, 3, 2)
plt.imshow(plt.imread(f'Data/Train/3/{view_dict[3]}'))
plt.title('3')

plt.subplot(1, 3, 3)
plt.imshow(plt.imread(f'Data/Train/drova/{view_dict[0]}'))
plt.title('Дрова')

plt.show()
```

```python
number_1 = len(os.listdir('Data/Train/1'))
number_3 = len(os.listdir('Data/Train/3'))
number_drova = len(os.listdir('Data/Train/drova'))

total_number = number_1 + number_3 + number_drova

plt.figure(figsize=(5, 5))
plt.grid()
plt.bar(['1', '3', 'Дрова'], [number_1 / total_number, number_3 / total_number, number_drova / total_number], width=0.3, color='purple')
plt.show()
```
